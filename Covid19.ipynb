```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid_df = pd.read_csv("covid19_Confirmed_dataset.csv")
covid_df.head()
```

```python
covid_df.shape
```

```python
covid_df.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
covid_df.head()
```

Aggreagate the rows by country

```python
corona_dataset_aggregate = covid_df.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregate.head()
```

```python
corona_dataset_aggregate.shape
```

```python
print(corona_dataset_aggregate.head())
print(corona_dataset_aggregate.dtypes)
```

```python
print(corona_dataset_aggregate.dtypes)
```

```python
countries=list(corona_dataset_aggregate.index)
```

```python
max_infection_rates = []

for c in countries:
    # Get only numeric values (e.g., daily cases) for the country
    data = corona_dataset_aggregate.loc[c].astype(str).str.replace(',', '')  # remove commas if present
    data = pd.to_numeric(data, errors='coerce')  # convert to numbers, invalid parsing will be set as NaN
    max_rate = data.diff().max()  # calculate the max daily increase
    max_infection_rates.append(max_rate)

# Now assign the result
corona_dataset_aggregate["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregate
```

create a new dataframe

```python
corona_data=pd.DataFrame(corona_dataset_aggregate["Max_infection_rates"])
```

```python
corona_data
```
